```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Set prior parameters for Beta(1,1)
alpha_prior = 1
beta_prior = 1

# Observed data: 1 success and 15 failures
successes = 1
failures = 15

# Update the posterior parameters
alpha_posterior = alpha_prior + successes
beta_posterior = beta_prior + failures

# Print the updated posterior parameters
print(f"Posterior Beta Distribution Parameters:")
print(f"Alpha (Successes): {alpha_posterior}")
print(f"Beta (Failures): {beta_posterior}")

# Generate x values for plotting (probabilities from 0 to 1)
x = np.linspace(0, 1, 100)

# Calculate the Beta PDF for the prior and posterior distributions
prior_pdf = beta.pdf(x, alpha_prior, beta_prior)
posterior_pdf = beta.pdf(x, alpha_posterior, beta_posterior)

# Plot the prior and posterior distributions
plt.figure(figsize=(10, 5))

# Plot the prior distribution Beta(1,1)
plt.subplot(1, 2, 1)
plt.plot(x, prior_pdf, 'blue', lw=2)
plt.fill_between(x, prior_pdf, color='blue', alpha=0.3)
plt.title('Prior Distribution Beta(1,1)')
plt.xlabel('p (Probability of Success)')
plt.ylabel('Density')

# Plot the posterior distribution Beta(2,16)
plt.subplot(1, 2, 2)
plt.plot(x, posterior_pdf, 'green', lw=2)
plt.fill_between(x, posterior_pdf, color='green', alpha=0.3)
plt.title('Posterior Distribution Beta(2,16)')
plt.xlabel('p (Probability of Success)')
plt.ylabel('Density')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Prior parameters (Beta(1,1))
alpha_prior = 1
beta_prior = 1

# Observed data: 1 success and 15 failures
successes = 1
failures = 15

# Posterior parameters (Beta(2,16))
alpha_posterior = alpha_prior + successes
beta_posterior = beta_prior + failures

# Define the likelihood function for Bernoulli trials
def likelihood(p, successes, failures):
    return (p ** successes) * ((1 - p) ** failures)

# Generate values for p (probabilities from 0 to 1)
p_values = np.linspace(0, 1, 100)

# Calculate the Beta PDF for prior and posterior
prior_pdf = beta.pdf(p_values, alpha_prior, beta_prior)
posterior_pdf = beta.pdf(p_values, alpha_posterior, beta_posterior)

# Calculate the likelihood function
likelihood_values = likelihood(p_values, successes, failures)

# Normalize likelihood values for better visualization
likelihood_values /= np.max(likelihood_values)

# Plotting with transparent background
plt.figure(figsize=(8, 6), facecolor='none')

# Plot the prior distribution (Beta(1,1))
plt.plot(p_values, prior_pdf, color='blue', lw=2, label='Prior Beta(1,1)')
plt.fill_between(p_values, prior_pdf, color='blue', alpha=0.3)

# Plot the likelihood function
plt.plot(p_values, likelihood_values, color='red', lw=2, label='Likelihood')
plt.fill_between(p_values, likelihood_values, color='red', alpha=0.3)

# Plot the posterior distribution (Beta(2,16))
plt.plot(p_values, posterior_pdf, color='green', lw=2, label='Posterior Beta(2,16)')
plt.fill_between(p_values, posterior_pdf, color='green', alpha=0.3)

# Add title and labels
plt.title('Prior, Likelihood, and Posterior Distributions', fontsize=14)
plt.xlabel('p (Probability of Success)', fontsize=12)
plt.ylabel('Density / Likelihood', fontsize=12)

# Add a legend
plt.legend(loc='upper left')

# Set transparent background
plt.gca().patch.set_alpha(0) # Makes the plot area transparent
plt.grid(True)

# Show the plot
plt.show()
```
